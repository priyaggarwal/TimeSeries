```python
#Creation of new Solar array with all numbers in string format converted to numpy.int64 format 
#All erroneous values were left as they were which allowed for data cleaning later

new_solar_array = []
for i in range(len(renewables_data['Solar'])):
    new_solar_array.append(pd.to_numeric(renewables_data.iloc[i]['Solar'], errors='ignore'))
    
print(type(new_solar_array[0]))
print(type(new_solar_array[22814]))
```

```python
#Checking length of each array post-processed to see if they would fit into the renewables_data dataframe

print("Checking length of Geothermal array: ", len(new_geothermal_array))
print("Checking length of Biomass array: ", len(new_biomass_array))
print("Checking length of Biogas array: ", len(new_biogas_array))
print("Checking length of Small Hydro array: ", len(new_smallHydro_array))
print("Checking length of Wind array: ", len(new_wind_array))
print("Checking length of Solar array: ", len(new_solar_array))
```

```python
#Dropping all of the previous feature columns since they contained numbers in string format instead of in numerical formats
#Also dropping useless columns created during datascraping and initial cleaning

renewables_data = renewables_data.drop('Geothermal', axis=1)
renewables_data = renewables_data.drop('Biomass', axis=1)
renewables_data = renewables_data.drop('Biogas', axis=1)
renewables_data = renewables_data.drop('Small Hydro', axis=1)
renewables_data = renewables_data.drop('Wind', axis=1)
renewables_data = renewables_data.drop('Solar', axis=1)
renewables_data = renewables_data.drop('Index', axis=1)
renewables_data = renewables_data.drop('Solar_num', axis=1)
renewables_data
```

```python
#Assigning new feature columns to be the post-processing arrays with numbers in proper numerical format
#Also contains erroneous values in their original format allowing for distinction and cleaning

renewables_data['Geothermal'] = pd.Series(new_geothermal_array)
renewables_data['Biomass'] = pd.Series(new_biomass_array)
renewables_data['Biogas'] = pd.Series(new_biogas_array)
renewables_data['Small Hydro'] = pd.Series(new_smallHydro_array)
renewables_data['Wind'] = pd.Series(new_wind_array)
renewables_data['Solar'] = pd.Series(new_solar_array)
```

```python
#Checking results of processing and adding to dataframe
type(renewables_data.iloc[0]['Geothermal'])
```

```python
#Cleaning of dataframe by removing erroneous values in two steps 
#First the values in the 'Date' column which were not strings (i.e. those which were NaN values are kept), dropping the erroneous ones
#Then the values in the Other columns which are in the format np.int64 are kept, dropping the string values which are erroneous

renewables_data_cleaned = renewables_data[renewables_data['Date'].map(type) != str]
renewables_data_cleaned = renewables_data_cleaned[renewables_data_cleaned['Geothermal'].map(type) == np.int64]
renewables_data_cleaned = renewables_data_cleaned[renewables_data_cleaned['Biogas'].map(type) == np.int64]
renewables_data_cleaned = renewables_data_cleaned[renewables_data_cleaned['Small Hydro'].map(type) == np.int64]
renewables_data_cleaned = renewables_data_cleaned[renewables_data_cleaned['Wind'].map(type) == np.int64]
renewables_data_cleaned
```

```python
#Dropping of the Date column which is full of NaN values erroneously
#Dropping of the first column which is unnamed and contains outdated indexing

renewables_data_cleaned = renewables_data_cleaned.drop('Date', axis=1)
renewables_data_cleaned = renewables_data_cleaned.drop('Unnamed: 0', axis = 1)
```
